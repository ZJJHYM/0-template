# Split Video and Audio

```python
import os
import subprocess
from subprocess import Popen, PIPE
```

```python
# train_5k audio

output_dir = '/home/tione/notebook/tagging/dataset/train_5k/audio'
data_dir = '/home/tione/notebook/algo-2021/dataset/videos/video_5k/train_5k'

process_dict = dict()

for i, item in enumerate(os.listdir(data_dir)):
    if i % 100 == 0:
        print(i)
    
    item_path = os.path.join(data_dir, item)
    item_name = item.split('.')[0]
    
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    out_audio = os.path.join(output_dir, f'{item_name}.m4a')
    
    cmd = ['ffmpeg', '-i', item_path, '-acodec', 'copy', '-vn', out_audio]
    
    temp_process = subprocess.Popen(cmd, stdout=subprocess.PIPE, text=True)
    process_dict[temp_process.pid] = temp_process
    
    while len(process_dict) >= 8:
        for item in process_dict:
            if process_dict[item].poll() is not None:
                process_dict.pop(item)
                break
```

# Extract Frames

```python
# train_5k

output_dir = '/home/tione/notebook/tagging/dataset/train_5k/frames'

process_dict = dict()

n_frames_every_second = 4

data_dir = '/home/tione/notebook/algo-2021/dataset/videos/video_5k/train_5k'


for i, item in enumerate(os.listdir(data_dir)):
    item_path = os.path.join(data_dir, item)
    item_name = item.split('.')[0]
    out_dir = os.path.join(output_dir, item_name)
    
    if not os.path.exists(out_dir):
        os.makedirs(out_dir)
    
    out_file = os.path.join(out_dir, '%04d.jpg')
    
    cmd = ['ffmpeg', '-i', item_path, '-r', str(n_frames_every_second), '-q:v', '2', '-f', 'image2', out_file]
    
    temp_process = subprocess.Popen(cmd, stdout=subprocess.PIPE, text=True)
    process_dict[temp_process.pid] = temp_process
    
    while len(process_dict) >= 8:
        for item in process_dict:
            if process_dict[item].poll() is not None:
                process_dict.pop(item)
                break
```
